```python
import re
```

```python
def get_prefix(email):
    email_pattern = re.compile("([\w|\d| ! # $ % & ' * + \- \/ = ? ^ _ ` . { | } ~]+)@.*",re.IGNORECASE)
    prefix = re.search(email_pattern, email)
    if prefix:
        return prefix.group(1)
    else:
        return "undefined"
```

```python
def get_prefix_class(email_list):
    prefix_class = {}
    for item in email_list:
        prefix = get_prefix(item['email'])
        class_ = item['contact_type'] 
        if prefix:
            if prefix not in prefix_class:
                prefix_class[prefix] = {class_ : 1}
            elif class_ in prefix_class[prefix]:
                prefix_class[prefix][class_] += 1
            else:
                prefix_class[prefix][class_] = 1
    return prefix_class
```

```python
def merge_dict(prefix_class, prefix_counter):
    for prefix,class_count_dict in prefix_class.items():
        
        if prefix in prefix_counter:  # if prefix as a key in prefix_counter
            # looping each class name and count in prefix class[prefix]
            for class_name, count in prefix_class[prefix].items():
                if class_name in prefix_counter[prefix]:
                    prefix_counter[prefix][class_name] += count
                else:
                    prefix_counter[prefix][class_name] = count
        else:
            prefix_counter[prefix] = prefix_class[prefix]
    return prefix_counter
```

```python
import json
```

```python
def prefix_class_counter(email_data):
    prefix_counter = {}
#     for email_list_str in email_data:
#         email_list = json.loads(email_list_str)  # convert the str into json obj [{}] is in a string format
    for email_item_str in email_data:
        email_item = json.loads(email_item_str)
        prefix_class = get_prefix_class(email_item)
        prefix_counter = merge_dict(prefix_class, prefix_counter)
#     break
    return prefix_counter
```

# assigining class to the prefix

```python
def get_most_probable_class(dict_obj):
    highest_count = 0
    class_ = ""
    for key,value in dict_obj.items():
        if value > highest_count:
            class_ = key
            highest_count = value
    return class_
```

```python
def highest_class_prefix(prefix_class_ds): #returns dictionary of each prefix with corresponding class with highest value
    prefix_class_list = []
    for prefix in prefix_class_ds:
        prefix_class = {}
        prefix_class['prefix'] = prefix
        prefix_class['class'] = get_most_probable_class(prefix_class_ds[prefix])
        prefix_class_list.append(prefix_class)
    return prefix_class_list
```

```python
prefix_class = highest_class_prefix(prefix_class_ds)
```
